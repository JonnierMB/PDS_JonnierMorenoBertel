```python
#Muestreo de una seÃ±al
import numpy as np
import matplotlib.pyplot as plt

# Definimos los parÃ¡metros de la seÃ±al
frecuencia = 1 # Hz
amplitud = 1 # unidad
fase = 0 # radianes
duraciÃ³n = 1 # segundos
tasa_de_muestreo = 100 # muestras por segundo

# Generamos la seÃ±al analÃ³gica sinusoidal
t = np.linspace(0, duraciÃ³n, int(duraciÃ³n * tasa_de_muestreo), endpoint=False)
x_analÃ³gica = amplitud * np.sin(2 * np.pi * frecuencia * t + fase)

# Tomamos muestras de la seÃ±al analÃ³gica en los tiempos definidos por la frecuencia de muestreo
x_discreta = amplitud * np.sin(2 * np.pi * frecuencia * t + fase)

# Graficamos la seÃ±al analÃ³gica y discreta
fig, ax = plt.subplots(2, 1, figsize=(8, 6))
ax[0].plot(t, x_analÃ³gica)
ax[0].set_xlabel('Tiempo (s)')
ax[0].set_ylabel('Amplitud')
ax[0].set_title('SeÃ±al analÃ³gica sinusoidal de 1 Hz')
ax[1].stem(t, x_discreta)
ax[1].set_xlabel('Tiempo (s)')
ax[1].set_ylabel('Amplitud')
ax[1].set_title('SeÃ±al discreta sinusoidal de 1 Hz')
plt.tight_layout()
plt.show()
```

```python
#Recuerar seÃ±al analoga a partir de una seÃ±al digital
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Definimos los parÃ¡metros de la seÃ±al
frecuencia = 1 # Hz
amplitud = 1 # unidad
fase = 0 # radianes
duraciÃ³n = 1 # segundos
tasa_de_muestreo = 100 # muestras por segundo

# Generamos la seÃ±al analÃ³gica sinusoidal
t = np.linspace(0, duraciÃ³n, int(duraciÃ³n * tasa_de_muestreo), endpoint=False)
x_analÃ³gica = amplitud * np.sin(2 * np.pi * frecuencia * t + fase)

# Tomamos muestras de la seÃ±al analÃ³gica en los tiempos definidos por la frecuencia de muestreo
x_discreta = amplitud * np.sin(2 * np.pi * frecuencia * t + fase)

# Aplicamos un filtro paso bajo para recuperar la seÃ±al analÃ³gica original
frecuencia_de_corte = 5 # Hz
orden_del_filtro = 6
b, a = signal.butter(orden_del_filtro, frecuencia_de_corte / (tasa_de_muestreo / 2), 'lowpass')
x_recuperada = signal.filtfilt(b, a, x_discreta)

# Graficamos la seÃ±al analÃ³gica, la seÃ±al discreta y la seÃ±al recuperada
fig, ax = plt.subplots(3, 1, figsize=(8, 10))
ax[0].plot(t, x_analÃ³gica)
ax[0].set_xlabel('Tiempo (s)')
ax[0].set_ylabel('Amplitud')
ax[0].set_title('SeÃ±al analÃ³gica sinusoidal de 1 Hz')
ax[1].stem(t, x_discreta)
ax[1].set_xlabel('Tiempo (s)')
ax[1].set_ylabel('Amplitud')
ax[1].set_title('SeÃ±al discreta sinusoidal de 1 Hz')
ax[2].plot(t, x_recuperada)
ax[2].set_xlabel('Tiempo (s)')
ax[2].set_ylabel('Amplitud')
ax[2].set_title('SeÃ±al recuperada por filtro paso bajo')
plt.tight_layout()
plt.show()
```

```python
#RepresentaciÃ³n de funciones fundamentales desplazadas 
import numpy as np
import matplotlib.pyplot as plt

# Crear un vector de tiempo
t = np.arange(-1, 10)

# Crear una funciÃ³n escalÃ³n desplazada 5 muestras
u = np.zeros(len(t))
u[t >= 5] = 1

# Crear una funciÃ³n rampa desplazada 5 muestras
r = np.zeros(len(t))
r[t >= 5] = t[t >= 5] - 5

# Crear una funciÃ³n impulso desplazada 5 muestras
d = np.zeros(len(t))
d[t == 5] = 1

# Graficar las funciones
fig, ax = plt.subplots(3, 1, figsize=(8, 6))

ax[0].stem(t, u)
ax[0].set_title('FunciÃ³n escalÃ³n')

ax[1].stem(t, r)
ax[1].set_title('FunciÃ³n rampa')

ax[2].stem(t, d)
ax[2].set_title('FunciÃ³n impulso')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar seÃ±al escalÃ³n
t = np.linspace(0, 10, 15)
x = np.zeros_like(t)
x[t >= 3] = 1

# Aplicar filtro de media mÃ³vil de 3 puntos
filtro = np.ones(3) / 3
y = np.convolve(x, filtro, mode='valid')

# Graficar seÃ±al original
fig, ax = plt.subplots(figsize=(8, 6))
ax.stem(t, x)
ax.set_title('SeÃ±al escalÃ³n original')
ax.set_xlabel('Tiempo')
ax.set_ylabel('Amplitud')
plt.show()

# Graficar seÃ±al filtrada
fig, ax = plt.subplots(figsize=(8, 6))
ax.stem(t[1:-1], y)
ax.set_title('SeÃ±al escalÃ³n filtrada')
ax.set_xlabel('Tiempo')
ax.set_ylabel('Amplitud')
plt.show()
```

```python
#ConvoluciÃ³n
import numpy as np

def escalon(t):
    return np.piecewise(t, t>=0, [1, 0])

def exponencial(t):
    return np.exp(-t)

import matplotlib.pyplot as plt

# Definimos el tiempo y el paso de tiempo
t = np.linspace(0, 10, 10)
dt = t[1] - t[0]

# Creamos las seÃ±ales
esc = escalon(t)
exp = exponencial(t)


conv = np.convolve(escalon(t), exponencial(t), mode='full')*dt


# Graficamos las seÃ±ales por separado
plt.stem(t, esc, linefmt='blue', markerfmt='bo', label='EscalÃ³n')
plt.stem(t, exp, linefmt='red', markerfmt='ro', label='Exponencial')
plt.legend()
plt.show()

# Graficamos la convoluciÃ³n
plt.stem(np.linspace(0, 20, len(conv)), conv, linefmt='green', markerfmt='go', label='ConvoluciÃ³n')
plt.legend()
plt.show()
```

```python
# Sistema estable polo al interior de circunferencia unitaria
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sp
def zeropoles(b, a=1):
    # Genera una grafica con los polos y ceros de la funciÃ³n de transferencia
    # y la circunferencia unitaria.
    w,h = sp.freqz(b, a)
    sys1 = sp.lti(b, a)
    ang = np.arange(0.0, 2*np.pi, 0.01)
    xp = np.cos(ang)
    yp = np.sin(ang)
    plt.figure()
    plt.plot(xp, yp, '--')
    plt.plot(sys1.zeros.real, sys1.zeros.imag, 'o')
    plt.plot(sys1.poles.real, sys1.poles.imag, 'x')
    plt.grid()
    plt.show()

    
b = [1] # Numerados
a = [1, -1/2]   # Denominador
zeropoles(b, a)
# 1 / (z-1/2) ->1/z (1-1/2z)
#ROC: |Z| > 0,5
```
